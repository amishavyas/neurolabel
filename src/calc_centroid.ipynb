# Calculate the Centroid for a Given Parcellation

```python
import numpy as np
import pandas as pd
from nilearn.datasets import fetch_atlas_schaefer_2018
from nilearn.plotting import find_parcellation_cut_coords
from nilearn.plotting import find_xyz_cut_coords
from nilearn import image
import nibabel as nib
import matplotlib.pyplot as plt
from utils import *
from scipy.ndimage import center_of_mass
```

```python
neurosynth_parcellation_path = "~/Code/GitHub/neurolabel/parcellations/neurosynth/Neurosynth_Parcellation_k100.nii.gz"
parcellation_path = "~/Code/GitHub/neurolabel/parcellations/Schaefer2018_400Parcels_17Networks_order_FSLMNI152_1mm.nii.gz"
img = nib.load(neurosynth_parcellation_path)
data = img.get_fdata()
affine=img.affine

unique_labels = np.round(np.unique(data))
unique_labels = unique_labels[unique_labels !=0]

#print(unique_labels)
xyz_coords = []
for label in unique_labels:
    print(label)
    mass_calc_df = np.round(data)==label
    centroid = center_of_mass(mass_calc_df) # (data == label)

    mni_x, mni_y, mni_z = image.coord_transform(centroid[0], centroid[1], centroid[2], affine)
    x = np.round(mni_x)
    y = np.round(mni_y)
    z = np.round(mni_z)
    output = [x,y,z]
    print(output)
    xyz_coords.append(output)

output_df = pd.DataFrame(xyz_coords, columns=["x", "y", "z"])                     
output_df.insert(0,"parcel_num", unique_labels.astype(float))

output_df.to_csv("~/Code/GitHub/neurolabel/parcellations/neurosynth/neurosynth100_centroids.csv", index=False)
```
