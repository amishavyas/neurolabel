## atlas_overlap

```python
x = 10
neurosynth100 = nlab.load_neurosynth("neurosynth100")
parcel_x_overlap_df = nlab.atlas_overlap(target_atlas=neurosynth100, parcel_num=x, label_atlas="schaefer", intersect_only=True)
parcel_x_overlap_df
```

## plot_anat

```python
parcel_x_anat_barplot = nlab.plot_anat(overlap_df=parcel_x_overlap_df, top_n=11, bar_color="#567660")

# Uncomment the line below to save out and view the example barplot
# parcel_x_anat_barplot.savefig("example_anat_barplot.PNG")
```
